```python
model_dir=r"D:\ALL PROJECTS\Face-anti-spoofing-system\saved_models\face_antispoofing_v3_224.keras"
import tensorflow as tf
model=tf.keras.models.load_model(model_dir)
```

```python
img_test_path=r"D:\ALL PROJECTS\Face-anti-spoofing-system\eval_and_track\datasets\images.jpeg"
import cv2
import numpy as np
img=cv2.imread(img_test_path)
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img=cv2.resize(img,(224,224))
img=img.astype('float32')
img=np.expand_dims(img,axis=0)
print(img.shape)
```

```python
results=model.predict(img)[0][0]
label="Real" if results>=0.5 else "Spoof"

mlflow.log_metric("score",results)
mlflow.log_param("Label",label)
mlflow.log_artifact(img_test_path)

```

```python
import os
dataset_path=os.path.join(os.getcwd(),"..","data","Detectedface")
```

```python
import os
import cv2
import numpy as np
from pathlib import Path
```

```python
print(os.path.exists(dataset_path))
```

```python
images_paths=[]
labels=[]

```

```python
reading=os.walk(dataset_path)
for root,dirs,files in reading:
    for i in files:
        if i.endswith('.jpg'):
            if "ClientFace" in root:
                labels.append(0)
            elif "ImposterFace" in root:
                labels.append(1)
            images_paths.append(os.path.join(root, i))
print(len(images_paths))
print(len(labels))
```

```python
outputs=[]
for i in images_paths:
    img=cv2.imread(i)
    img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    img=cv2.resize(img,(224,224))
    img=img.astype('float32')
    img=np.expand_dims(img,axis=0)
    ypreds=model.predict(img)
    if ypreds>=0.6:
        outputs.append(0)
    else:
        outputs.append(1)
    
    
```
