```python
import mlflow
```

```python
mlflow.set_tracking_uri("http://127.0.0.1:5000/")
```

```python
mlflow.set_experiment("Initial Testing")
```

```python
mlflow.start_run()
```

```python
model_dir=r"D:\ALL PROJECTS\Face-anti-spoofing-system\saved_models\face_antispoofing_v3_224.keras"
import tensorflow as tf
model=tf.keras.models.load_model(model_dir)
```

```python
img_test_path=r"D:\ALL PROJECTS\Face-anti-spoofing-system\eval_and_track\datasets\images.jpeg"
import cv2
import numpy as np
img=cv2.imread(img_test_path)
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img=cv2.resize(img,(224,224))
img=img.astype('float32')
img=np.expand_dims(img,axis=0)
print(img.shape)
```

```python
results=model.predict(img)[0][0]
label="Real" if results>=0.5 else "Spoof"

mlflow.log_metric("score",results)
mlflow.log_param("Label",label)
mlflow.log_artifact(img_test_path)

```
